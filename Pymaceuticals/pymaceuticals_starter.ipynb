```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_df, clinical_data_df, on="Mouse ID", how="left")

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_by_drug_time_gr = combined_data.groupby(["Drug", "Timepoint"])
gr = tumor_by_drug_time_gr["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
avg_tumor_size_by_drug = pd.DataFrame(gr).reset_index()

# Preview DataFrame
avg_tumor_size_by_drug.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
std_err_gr = tumor_by_drug_time_gr["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
std_err_df = pd.DataFrame(std_err_gr).reset_index()

# Preview DataFrame
std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_vol_pivot = avg_tumor_size_by_drug.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_vol_pivot.head()
```

```python
#Data for charts 
x_axis = np.arange(0, 50, 5)
mouse_count = 25
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo", "Ramicane"]
markers = ["o", "^", "s", "d", "x"]
colors = ["red", "blue", "green", "black", "purple"]
start_day = 0
end_day = 45

#### I decided to inlcude Ramicane in the charts as well after I noticed it had a high reduction in tumor volume 
```

```python
# Generate the Plot (with Error Bars)
metric = "Tumor Volume (mm3)"

fig1, ax1 = plt.subplots()

for d in drugs:
    drug_data = avg_tumor_size_by_drug[avg_tumor_size_by_drug["Drug"]==d][metric].values
    drug_err = std_err_df[std_err_df["Drug"]==d][metric].values
    ax1.errorbar(x_axis, drug_data, drug_err, color=colors[drugs.index(d)], marker=markers[drugs.index(d)], linewidth=0.5, label=d)
  
plt.title("Tumor Response to Treatment")
plt.ylabel(metric)
plt.xlabel("Time (Days)")
ax1.legend()
ax1.yaxis.grid()

# Save the Figure
plt.savefig("tumor_response_to_treatment.jpg")

#Show the Figure
plt.show()


```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatis_by_drug_time_gr = combined_data.groupby(["Drug", "Timepoint"])
gr = metastatis_by_drug_time_gr["Metastatic Sites"].mean()

# Convert to DataFrame
avg_metastatis_by_drug = pd.DataFrame(gr).reset_index()

# Preview DataFrame
avg_metastatis_by_drug.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_err_gr = metastatis_by_drug_time_gr["Metastatic Sites"].sem()

# Convert to DataFrame
std_err_df = pd.DataFrame(std_err_gr).reset_index()

# Preview DataFrame
std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastasis_pivot = avg_metastatis_by_drug.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
metastasis_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
metric = "Metastatic Sites"

fig2, ax2 = plt.subplots()

for d in drugs:
    drug_data = avg_metastatis_by_drug[avg_metastatis_by_drug["Drug"]==d][metric].values
    drug_err = std_err_df[std_err_df["Drug"]==d][metric].values
    ax2.errorbar(x_axis, drug_data, drug_err, color=colors[drugs.index(d)], marker=markers[drugs.index(d)], linewidth=0.5, label=d)
    
plt.title("Metastatic Spread During Treatment")
plt.ylabel(metric)
plt.xlabel("Treatment Duration (Days)")
ax2.legend()
ax2.yaxis.grid()

# Save the Figure
plt.savefig("metastatic_spread_during_treatment.jpg")

# Show the Figure
plt.show()

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_by_drug_time_gr = combined_data.groupby(["Drug", "Timepoint"])
gr = survival_by_drug_time_gr["Mouse ID"].count()

# Convert to DataFrame
avg_survival_by_drug = pd.DataFrame(gr).reset_index()

# Preview DataFrame
avg_survival_by_drug.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_pivot = avg_survival_by_drug.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview that Reformatting worked
survival_pivot.head(20)
```

```python
# Generate the Plot (Accounting for percentages)
metric = "Mouse ID"

fig3, ax3 = plt.subplots()

for d in drugs:
    drug_data = avg_survival_by_drug[avg_survival_by_drug["Drug"]==d][metric].values/mouse_count*100
    ax3.plot(x_axis, drug_data, color=colors[drugs.index(d)], marker=markers[drugs.index(d)], linewidth=0.5, label=d)
    
plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Treatment Duration (Days)")
ax3.legend()
ax3.grid()

# Save the Figure
plt.savefig("survival_rate_per_treatment.jpg")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
start = avg_tumor_size_by_drug[avg_tumor_size_by_drug["Timepoint"]==start_day][["Drug", "Tumor Volume (mm3)"]]
end = avg_tumor_size_by_drug[avg_tumor_size_by_drug["Timepoint"]==end_day][["Drug", "Tumor Volume (mm3)"]]

diff = (end["Tumor Volume (mm3)"].values - start["Tumor Volume (mm3)"].values)/start["Tumor Volume (mm3)"].values*100

# Display the data to confirm
drug_effectiveness = pd.DataFrame()

drug_effectiveness["Drug"] = start["Drug"]
drug_effectiveness["Tumor Reduction (%)"] = diff

drug_effectiveness_limited = drug_effectiveness[drug_effectiveness["Drug"].isin(drugs)]

drug_effectiveness_limited

### not sure why Ramicane is not part of the comparison, it seems to be an effective drug...
### I think we need to make sure we compare the performance of other drugs against it as well
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
# Call functions to implement the function calls
# Save the Figure
# Show the Figure

bar_colors = ['green' if (x < 0) else 'red' for x in drug_effectiveness_limited["Tumor Reduction (%)"]]
ticks = np.arange(len(drugs))
plt.bar(ticks, drug_effectiveness_limited["Tumor Reduction (%)"],  color=bar_colors, align='edge', width=1)
plt.xticks(ticks, drug_effectiveness_limited["Drug"], rotation=45)
plt.yticks(np.arange(-40, 80, 20))
plt.ylim(-30, 70)
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")
plt.grid()

# Save the Figure
plt.savefig("percent_change_per_drug.jpg")

plt.show()
```

# Observations
- Capomulin and Ramicane drugs look to be the most effective drugs having positive outlook results in all three categories: Tumor Growth, Metastatis Spread, Survival Rate
- Capomulin and Ramicane had highest reduction in tumor size, reducing the size of the tumor by 19% and 22% respectively in only 45 days, while other drugs showed no reduction at all and the tumor continued to grow rapidly
- Capomulin and Ramicane drugs also show the lowest metastatic spread. The spread was 2X-3X lower compared to other drugs
- Capomulin and Ramicane also showed positive results on the survival rate with about 80% survival rate comparing to other drugs which had a survival rate between 40%-50%
